```python
#download BeautifulSoup library for Web Scraping
import bs4 as bs  
import urllib.request  
import re

scraped_data = urllib.request.urlopen('https://en.wikipedia.org/wiki/Jawaharlal_Nehru')  #make url request

article = scraped_data.read()

parsed_article = bs.BeautifulSoup(article,'lxml')
paragraphs = parsed_article.find_all('p') #get all p tags
article_text = ""

for p in paragraphs:  
    article_text += p.text #cancatinate all p tags data in article_text
```

```python
#use this block to read file on the computer
import re

sample = open("C:\\Users\\rehan\\Desktop\\doc2.txt", "r") 
s = sample.read() 
  
# Replaces escape character with space 
article_text = s.replace("\n", " ") 
article_text
```

```python
article_text = re.sub(r'\[[0-9]*\]', ' ', article_text)  
article_text = re.sub(r'\s+', ' ', article_text)  
formatted_article_text = re.sub('[^a-zA-Z]', ' ', article_text )  
formatted_article_text = re.sub(r'\s+', ' ', formatted_article_text)
```

```python
#create sentence tokens
sentence_list = nltk.sent_tokenize(article_text)
```

```python
#calculate frequency of each word
stopwords = nltk.corpus.stopwords.words('english')
word_frequencies = {}  
for word in nltk.word_tokenize(formatted_article_text):  
    if word not in stopwords:
        if word not in word_frequencies.keys():
            word_frequencies[word] = 1
        else:
            word_frequencies[word] += 1
stopwords
```

```python
word_frequencies
```

```python
#normalize the frequency dataset 
maximum_frequncy = max(word_frequencies.values())
for word in word_frequencies.keys():  
    word_frequencies[word] = (word_frequencies[word]/maximum_frequncy)
```

```python
#Calculate sentence score of each sentence
sentence_scores = {}  
for sent in sentence_list:  
    for word in nltk.word_tokenize(sent.lower()):
        if word in word_frequencies.keys():
            if len(sent.split(' ')) < 30: #30 repersents maz no of words a sentence can have
                if sent not in sentence_scores.keys():
                    sentence_scores[sent] = word_frequencies[word]
                else:
                    sentence_scores[sent] += word_frequencies[word]
```

```python
#generating the summary
import heapq
summary_sentences = heapq.nlargest(7, sentence_scores, key=sentence_scores.get)

summary = ' '.join(summary_sentences)  
print(summary)  
```
